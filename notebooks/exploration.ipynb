```python
import os
import requests
from bs4 import BeautifulSoup
from pathlib import Path

WD = Path(os.getcwd()).parent

DATA_DIR = WD / 'data'



def download_opinions(url, download_dir):
    # Fetch the content of the webpage
    response = requests.get(url)
    response.raise_for_status()  # Raises an error for bad responses

    # Use BeautifulSoup to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all links to opinion PDFs
    opinion_links = soup.find_all('a', href=True)

    for link in opinion_links:
        href = link['href']
        try:
            if href.endswith('.pdf'):  # Check if the link is to a PDF file
                pdf_url = f"https://www.supremecourt.gov{href}" if not href.startswith('http') else href

                print(f"pdf_url - {pdf_url}")
                pdf_response = requests.get(pdf_url)
                pdf_name = href.split('/')[-1]
                
                # Save the PDF to the specified directory
                with open(os.path.join(download_dir, pdf_name), 'wb') as pdf_file:
                    pdf_file.write(pdf_response.content)
                print(f"Downloaded {pdf_name}")
        except Exception as e:
            print(f"Error downloading {pdf_url}: {e}")
            continue



year = 22
#year = 24
# URL of the Supreme Court opinions page
url = f"https://www.supremecourt.gov/opinions/slipopinion/{year}"

# Directory to save downloaded opinions
download_dir = DATA_DIR/ f"./opinions_{year}"
os.makedirs(download_dir, exist_ok=True)
# Call the function
download_opinions(url, download_dir)
```
